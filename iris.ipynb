```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(data=np.c_[iris['data'], iris['target']],
                  columns=iris['feature_names'] + ['target'])
```

```python
df
```

```python
print(df.isnull().sum())
```

```python
print(df.head())
print(df.describe())
X = df.drop('target', axis=1)
y = df['target']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
sns.pairplot(df, hue='target', palette='Set1')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=X)
plt.title('Feature Distribution (Boxplot)')
plt.xlabel('Features')
plt.ylabel('Values')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,y,test_size=0.3,random_state=7)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=7).fit(X_train,Y_train)
```

```python
pre = model.predict(X_test)
pre
```

```python
Y_test
```

```python
from sklearn import metrics
print("Train set Accuracy: ", metrics.accuracy_score(Y_train, model.predict(X_train)))
print("Test set Accuracy: ", metrics.accuracy_score(Y_test, pre))
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
Y_Prediction_train=model.predict(X_train)

mean_train=mean_absolute_error(Y_train, Y_Prediction_train)
mean_train
```
